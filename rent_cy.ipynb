```python
import pandas as pd
from pyecharts.charts import *
import pyecharts.options as opts
```

```python
rent = pd.read_csv('Users/shawnhua/Desktop/rent1.csv')
```

# 数据清洗

```python
rent.drop('renta_type',axis=1,inplace=True)
rent.head()
```

```python
rent.drop(rent.index[0],inplace=True)
```

```python
rent.tail()
```

```python
rent = rent.dropna(subset=['rent_type'])
del rent['_id']
rent.reset_index(drop=True,inplace=True)
rent
```

```python
rent.isnull().sum()
```

```python
# 修改城区信息
distric=[]
for each in rent['distric']:
    l=list(each)
    l[-2:]='区'
    str1=''.join(l)
    distric.append(str1)
rent['distric'] = distric

#修改城区区域信息
bc=[]
for each in rent['bc']:
    l=list(each)
    del l[-2:]
    str2=''.join(l)
    bc.append(str2)
rent['bc'] = bc


#修改小区信息
community=[]
for each in rent['community']:
    l=list(each)
    del l[-2:]
    str3=''.join(l)
    community.append(str3)
rent['community'] = community

#修改房屋面积
size=[]
for each in rent['size']:
    l=list(each)
    del l[0:3]
    del l[-1]
    str4=''.join(l)
    size.append(str4)
rent['size'] = size

#修改城市
city=[]
for each in rent['city']:
    l=list(each)
    del l[-3:]
    str4=''.join(l)
    city.append(str4)
rent['city'] = city

#租赁方式
rent_type=[]
for each in rent['rent_type']:
    l=list(each)
    str4=''.join(l[5:7])
    rent_type.append(str4)
rent['rent_type'] = rent_type

rent = rent[rent['rent_type'] != '未知']

#房屋类型
house_type=[]
for each in rent['house_type']:
    l=list(each)
    str4=''.join(l[5:9])
    house_type.append(str4)
rent['house_type'] = house_type
rent
```

```python
rent['size'] = rent['size'].astype('float')  # 转换数据类型
rent.duplicated().sum()  # 去重
rent.drop_duplicates(inplace=True)
rent['rpua'] = round(rent['rental']/rent['size'])  # 单位面积租金
rent.describe()
```

## 1.成渝核心区房价排名情况
   ### 重庆租金现状分析
   按单位面积月租金统计，重庆地区平均租金在25.21元/(㎡/月），其中渝中区租金最高，江渝北租金接近，位列第二第三。原因在于：从总体上看，这三个区域对应的GDP是重庆市最高的，经济创造价值高的区域对应的人口活跃度高，对应的房价也高。从各区域定位看，渝中地区是重庆市的历史人文母城、政治中心，也是地理几何中心；江渝北地区是重庆就业中心，也是两江新区的核心区域，两江新区是国家级新区，重庆最重要的经济增长极，其中江北地区是重庆金融行业的聚集地，对应的劳动人口需求量大、薪资高，支撑起了该区域的租金；而渝北区是重庆企业数量最多的聚居地，也是就业人口最多的地区，不过该地区的面积较为广阔，大部分地区还发展未成熟，因此平均房价仅位于第三，相信随着时间的推移，城区发展成熟，未来的租金水平将会上升。
   ### 成都租金现状分析
   按单位面积月租金统计，成都地区平均租金在30.20元/(㎡/月），其中高新区租金最高，武侯、锦江区租金相同，为第二第三。原因在于：高新区是成都的就业中心，区域交通成熟，包含天府软件园在内的众多提供就业机会的聚集地，是成都新增人口首选区域，就业人员平均薪资水平高，这些因素促使该区域租赁房价处于首位。而武侯、锦江区是成都的传统核心区，处于成都中心城区一圈，设施配套成熟、交通便利，又临近高新区，因此租金水平也较高。
   ### 对比分析
   成都地区租金高于重庆，包含成都薪资水平高于重庆，且成都地区净流入人口数量高，对租房的需求量更大等因素

```python
# 重庆、成都信息情况
cq_info = rent[rent['city']=='重庆']
cd_info = rent[rent['city']=='成都']

# 成渝核心区租金情况
re_dis_cq = cq_info.groupby('distric')['rpua'].mean().sort_values(ascending=False).to_dict()
re_dis_cd = cd_info.groupby('distric')['rpua'].mean().sort_values(ascending=False).to_dict()
dis_cq = list(re_dis_cq.keys())
rental_cq = list(re_dis_cq.values())
dis_cd = list(re_dis_cd.keys())
rental_cd = list(re_dis_cd.values())

#成渝各区租房房价排名的可视化
l_cq,l_cd = [],[]
for i in rental_cq:
    l_cq.append(round(i))
for i in rental_cd:
    l_cd.append(round(i))
bar1 = (Bar()
       .add_xaxis(dis_cq)
       .add_yaxis('重庆',l_cq)
       .set_global_opts(title_opts=opts.TitleOpts(title='成渝各区租房房价排名（元/(㎡/月）'))
       .set_series_opts(
        label_opts=opts.LabelOpts(is_show=True),
        markline_opts=opts.MarkLineOpts(
            data=[
                opts.MarkLineItem(type_="min", name="最小值"),
                opts.MarkLineItem(type_="max", name="最大值"),
                opts.MarkLineItem(type_="average", name="平均值"),
            ]
        ),
    ))
bar2 = (Bar(init_opts=opts.InitOpts(theme='dark'))
       .add_xaxis(dis_cd)
       .add_yaxis('成都',l_cd)
       .set_series_opts(
        label_opts=opts.LabelOpts(is_show=True),
        markline_opts=opts.MarkLineOpts(
            data=[
                opts.MarkLineItem(type_="min", name="最小值"),
                opts.MarkLineItem(type_="max", name="最大值"),
                opts.MarkLineItem(type_="average", name="平均值"),
            ]
        ),
    )      
       )
page = (
    Page()
    .add(bar1,bar2)
    )
page.render_notebook()

```

## 3 户型和面积分布情况
   ### 户型分布情况
   考虑到不同租赁人对房屋户型的需求不同，本人通过对成渝地区核心区租房数据的户型按照**整租和合租**两个租赁类型进行分组统计，展示户型数量总体中占比较高的类型，用Pyecharts饼图展示得到户型分布图。
   统计发现，整租类型的租赁人，其中受欢迎的租房类型为3室2厅、2室1厅、1室1厅。其中选择3室2厅和2室1厅的人员理应为家庭，占比43.5%。选择1室1厅多为未婚，占比13.15%，两者占比超50%。合租类型的租赁人，其中受欢迎的租房类型为4室1厅、5室1厅和3室1厅，和占比超过80%，究其原因离不开房价更低。
